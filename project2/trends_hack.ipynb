```python
import pandas as pd
import requests as rq
import lxml as xl
import pytrends
from pytrends.request import TrendReq
pytrends = TrendReq()
```

```python
topic_ls = ["guns", "germs", "recreation"]
gun_ls = ["ammunition", "gun stores", "ar-15"]
germs_ls = ["hand sanitizer", "bleach", "toilet paper"]
tv_ls = ["netflix", "weed", "beer"]
```

```python
my_req = TrendReq(hl='en-US', tz=360)
my_req.build_payload(topic_ls, timeframe = 'today 3-m')
df_2019 = my_req.interest_over_time()
df_2019.head()
```

```python
#df_2019 = df_2019.drop("isPartial", axis =1)
df_2020 = df_2019
df_2020.to_csv('goog_csvs/topics_2020.csv')
```

```python
my_req = TrendReq(hl='en-US', tz=360)
my_req.build_payload(gun_ls, timeframe = 'today 3-m')
df_gun_2020 = my_req.interest_over_time()
df_gun_2020 = df_gun_2020.drop("isPartial", axis =1)
df_gun_2020.head()
```

```python
df_gun_2020.to_csv('goog_csvs/guns_2020.csv')
```

```python
my_req = TrendReq(hl='en-US', tz=360)
my_req.build_payload(germs_ls, timeframe = 'today 3-m')
df_germ_2020 = my_req.interest_over_time()
df_germ_2020.head()
```

```python
df_germ_2020 = df_germ_2020.drop("isPartial", axis = 1)
df_germ_2020.to_csv("goog_csvs/germs_2020.csv")
```

```python
my_req = TrendReq(hl='en-US', tz=360)
my_req.build_payload(tv_ls, timeframe = 'today 3-m')
df_tv_2020 = my_req.interest_over_time()
df_tv_2020.head()
```
