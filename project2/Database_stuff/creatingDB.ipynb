```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///trends.sqlite', echo=False)
```

```python
import pandas as pd
```

```python
trends2019 = pd.read_csv("../goog_csvs/2019_data.csv")
trends2019 = trends2019[['date', 'guns', 'virus', 'boredom', 'ammo', 'gun stores','ar15', 'disinfectant', 'bleach',
                        'toilet paper', 'netflix', 'porn', 'diy']]
trends2019.to_sql('trends2019', con=engine)
```

```python
trends2020 = pd.read_csv("../goog_csvs/2020_data.csv")
trends2020 = trends2020[['date', 'guns', 'virus', 'boredom', 'ammo', 'gun stores', 'ar15', 'disinfectant', 'bleach',
                        'toilet paper', 'netflix', 'porn', 'diy']]
trends2020.to_sql('trends2020', con=engine)
```

```python
netflixTweets = pd.read_csv('../tweet_csvs/netflix_tweets.csv')
netflixTweets.to_sql('netflix_tweet', con=engine)
```

```python
cdcTweet = pd.read_csv('../tweet_csvs/CDCEmergency_tweets.csv')
cdcTweet.to_sql('cdcTweet', con=engine)
```
