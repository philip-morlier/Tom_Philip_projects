```python
import pandas as pd
import requests as rq
import lxml as xl
import pytrends
from pytrends.request import TrendReq
import numpy
import datetime
pytrends = TrendReq()
```

```python
topic_ls = [["guns"], ["virus"], ["boredom"]]
gun_ls = [["ammo"], ["gun stores"], ["ar15"]]
germs_ls = [["disinfectant"], ["bleach"], ["toilet paper"]]
tv_ls = [["netflix"], ["porn"], ["diy"]]
```

```python
k = 0
df_topic = pd.DataFrame(columns = ["guns", "virus", "boredom"])
for i in topic_ls:
    my_req = TrendReq(hl='en-US', tz=360)
    my_req.build_payload(topic_ls[k], timeframe = 'today 5-y')
    df_holder = my_req.interest_over_time()
    df_topic[i[0]] = df_holder[i[0]]
    k = k + 1
df_topic
```

```python
k = 0
df_gun = pd.DataFrame(columns = ["ammo", "gun stores", "ar15"])
for i in gun_ls:
    my_req = TrendReq(hl='en-US', tz=360)
    my_req.build_payload(gun_ls[k], timeframe = 'today 5-y')
    df_holder = my_req.interest_over_time()
    df_gun[i[0]] = df_holder[i[0]]
    k = k + 1
df_gun
```

```python
k = 0
df_germs = pd.DataFrame(columns = ["disinfectant", "bleach", "toilet paper"])
for i in germs_ls:
    my_req = TrendReq(hl='en-US', tz=360)
    my_req.build_payload(germs_ls[k], timeframe = 'today 5-y')
    df_holder = my_req.interest_over_time()
    df_germs[i[0]] = df_holder[i[0]]
    k = k + 1
df_germs
```

```python
k = 0
df_tv = pd.DataFrame(columns = ["netflix", "porn", "diy"])
for i in tv_ls:
    my_req = TrendReq(hl='en-US', tz=360)
    my_req.build_payload(tv_ls[k], timeframe = 'today 5-y')
    df_holder = my_req.interest_over_time()
    df_tv[i[0]] = df_holder[i[0]]
    k = k + 1
df_tv
```

```python
df_topic = df_topic.reset_index()
df_gun = df_gun.reset_index()
df_germs = df_germs.reset_index()
df_tv = df_tv.reset_index()
```

```python
x = datetime.datetime(2020, 4, 1)
y = datetime.datetime(2019, 4, 1)
df_topic_20 = df_topic
df_topic_19 = df_topic
for index, inner in df_topic.iterrows():
    if inner["date"].year != x.year:
        df_topic_20 = df_topic_20.drop(index)
    elif inner["date"].month > x.month:
        df_topic_20 = df_topic_20.drop(index)
        
    if inner["date"].year != y.year:
        df_topic_19 = df_topic_19.drop(index)
    elif inner["date"].month > y.month:
        df_topic_19 = df_topic_19.drop(index)
```

```python
x = datetime.datetime(2020, 4, 1)
y = datetime.datetime(2019, 4, 1)
df_gun_20 = df_gun
df_gun_19 = df_gun
for index, inner in df_gun.iterrows():
    if inner["date"].year != x.year:
        df_gun_20 = df_gun_20.drop(index)
    elif inner["date"].month > x.month:
        df_gun_20 = df_gun_20.drop(index)
        
    if inner["date"].year != y.year:
        df_gun_19 = df_gun_19.drop(index)
    elif inner["date"].month > y.month:
        df_gun_19 = df_gun_19.drop(index)
```

```python
x = datetime.datetime(2020, 4, 1)
y = datetime.datetime(2019, 4, 1)
df_germs_20 = df_germs
df_germs_19 = df_germs
for index, inner in df_germs.iterrows():
    if inner["date"].year != x.year:
        df_germs_20 = df_germs_20.drop(index)
    elif inner["date"].month > x.month:
        df_germs_20 = df_germs_20.drop(index)
        
    if inner["date"].year != y.year:
        df_germs_19 = df_germs_19.drop(index)
    elif inner["date"].month > y.month:
        df_germs_19 = df_germs_19.drop(index)
```

```python
x = datetime.datetime(2020, 4, 1)
y = datetime.datetime(2019, 4, 1)
df_tv_20 = df_tv
df_tv_19 = df_tv
for index, inner in df_tv.iterrows():
    if inner["date"].year != x.year:
        df_tv_20 = df_tv_20.drop(index)
    elif inner["date"].month > x.month:
        df_tv_20 = df_tv_20.drop(index)
        
    if inner["date"].year != y.year:
        df_tv_19 = df_tv_19.drop(index)
    elif inner["date"].month > y.month:
        df_tv_19 = df_tv_19.drop(index)
```

```python
df_topic_20.to_csv("goog_csvs/topics_2020.csv")
df_topic_19.to_csv("goog_csvs/2019_topic.csv")
df_gun_20.to_csv("goog_csvs/guns_2020.csv")
df_gun_19.to_csv("goog_csvs/2019_guns.csv")
df_germs_20.to_csv("goog_csvs/germs_2020.csv")
df_germs_19.to_csv("goog_csvs/2019_germs.csv")
df_tv_20.to_csv("goog_csvs/rec_2020.csv")
df_tv_19.to_csv("goog_csvs/2019_rec.csv")
```
